```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:/Users/User/Documents/data_analysis_wit_python1/mod2.csv'
df1 = pd.read_csv(file_path)
df = df1
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df
```

```python
# We remove the Unnamed: column
df.drop('Unnamed: 0',axis= 1, inplace= True)
```

```python
# Replacing missing values for each column
avgsym = df['symboling'].max()
df['symboling'] = df['symboling'].replace(np.NaN, avgsym)
df['symboling']
```

```python
mississ = df.isnull()
for columns in mississ.columns.values.tolist():
    print(mississ[columns].value_counts())
```

```python
avgnorm = df['normalized-losses'].mean()
df['normalized-losses'] = df['normalized-losses'].replace(np.NaN,avgnorm)
```

```python
k = df['normalized-losses']. isnull()
k.value_counts()
```

```python
avgasp = df['aspiration'].max()
df['aspiration'] = df['aspiration'].replace(np.NaN,avgasp)
```

```python
avgpr = df['price'].mean()
df['price'] = df['price'].replace(np.NaN,avgpr)
```

```python
df
```

```python
#  data formating
df['city-mpg'] = 235/df['city-mpg']
df.rename(columns = {'city-mpg':'city-L/100km'}, inplace= True)
```

```python
df
```

```python
df['price'].dtypes
```

```python
df['price'] = df['price'].astype('int')
```

```python
# Data Binning
bins= np.linspace(min(df['price']),max(df['price']),4)
group_names = {'Low', 'Medium', 'High'}
df['price-binned'] = pd.cut(df['price'], bins, labels = group_names)
df['price-binned'].value_counts()
```

```python
df
```

```python
# data normalization
li = df['length'] /df['length'].max()
```

```python
df['length'] = li
```

```python
dummy_variable_1=pd.get_dummies(df['fuel-type'])
dummy_variable_1.rename(columns={'diesel':'fuel-type-diesel', 'gas':'fuel-type-gas'},inplace=True)
df = pd.concat([df,dummy_variable_1], axis = 1)
```

```python
df
```

```python
# Dropping Fuel Type
df.drop(['fuel-type'], axis=1, inplace=True)
```
